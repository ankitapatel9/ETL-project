```python
# File to Load (Remember to Change These)
chicago_crime_data = "Resources/Chicago_Crimes_2012_to_2017.csv"
chicago_df = pd.read_csv(chicago_crime_data)
chicago_df
```

```python
chicago_15_data = chicago_df[chicago_df['Year']==2015]
chicago_15_data.dropna(how='any')
```

```python
new_data = chicago_15_data[['Date','Primary Type','Community Area']]

crime_transformed = new_data.rename(columns={
                                            "Date": "date_time",
                                            "Primary Type": "primary_type",
                                            "Community Area" : "community_area"
                                                         })
# crime_transformed.drop_duplicates("id", inplace=True)
# crime_transformed.set_index("id", inplace=True)
crime_transformed['date'] = pd.to_datetime(crime_transformed['date_time'])
crime_transformed['date'] = crime_transformed['date'].dt.date
crime_transformed.head()
```

```python
count_crime = crime_transformed[['primary_type','date','community_area']]
count_crime.head()
```

```python
group_crime = count_crime.groupby(['date','primary_type'])['primary_type'].count()
df = pd.DataFrame(group_crime)
df = df.reset_index(level = 'date')
df.columns = ['date','num_crimes']
crime_df1 =df.reset_index()
crime_df1.head()
#new_df = df.pivot( index='date',columns='primary_type', values='num_crimes')
#new_df
```

```python
temp_data = "Resources/temperature.csv"
temp_df = pd.read_csv(temp_data)
temp_df.head()
```

```python
temp_df.dropna()
```

```python
temp_df.columns
```

```python
temp_cols = temp_df[["datetime","Chicago"]]
temp_transformed = temp_cols.rename(columns={
                                            "datetime": "date_time",
                                            "Chicago": "chicago"
                                                         })
temp_transformed['date'] = pd.to_datetime(temp_transformed['date_time'])
temp_transformed['date'] = temp_transformed['date'].dt.date
temp_transformed.head()
# temp_df1 = temp_transformed.dropna()
# temp_df1.head()
```

```python
temp_transformed['year'] = pd.DatetimeIndex( temp_transformed['date']).year
temp_15_data = temp_transformed[temp_transformed['year']== 2015]
temp_15_data.head()
```

```python
group_by_temp = temp_15_data.groupby(['year','date'])['chicago'].mean()
group_by_temp
final_temp = pd.DataFrame(group_by_temp)
final_temp = final_temp.reset_index()
final_temp.head()
```

```python
def kelvin_to_fahr(k):
    """Convert Fahrenheit to Celsius
    
    Return Celsius conversion of input"""
    temp_f = (k -273.15) * 9 / 5 + 32
    return temp_f
```

```python
final_temp["chicago_f"] = kelvin_to_fahr(round(final_temp["chicago"]))
```

```python
final_temp.head()
```

```python
chicago_temp_data = final_temp[['date','chicago_f']]
chicago_temp_data.head()
```

```python
chicago_temp_data.reset_index(inplace = True)
chicago_temp_data.rename(columns={"index": "id"})
```
